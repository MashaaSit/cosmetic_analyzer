```python
import pandas as pd
```

```python
df1 = pd.read_csv('downloaded_data\1-200.csv')
df2 = pd.read_csv('downloaded_data\201-400.csv')
df3 = pd.read_csv('downloaded_data\401-600.csv')
```

```python
df = pd.concat([df1, df2, df3], axis=0)
```

```python
df.head()
```

```python
ingredients_list = {}
```

```python
for row in df['ingredients']:
    for item in row.split(', '):
        if item not in ingredients_list:
            ingredients_list[item] = 0
        ingredients_list[item] += 1
```

```python
sorted_dict = dict(sorted(ingredients_list.items(),key=lambda item: item[1], reverse=True))
sorted_dict
```

```python
len(sorted_dict)
```

```python
from itertools import islice  

res = dict(islice(sorted_dict.items(), 0,200)) 
```

```python
data = {'ingredient_name': res.keys()}
ingredients_df = pd.DataFrame.from_dict(data)
```

```python
ingredients_df.to_csv('ingredients_classification.csv')
```
